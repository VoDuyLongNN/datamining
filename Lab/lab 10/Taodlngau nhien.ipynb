```python
import numpy as np # thư viện tính toán toán học
import matplotlib.pyplot as plt # visualize data sử dụng đồ thị
from scipy.spatial.distance import cdist # Hỗ trợ tính khoảng cách
```

```python
means = [[2, 2], [9, 2], [4, 9]]
cov = [[2, 0], [0, 2]]
n_samples = 500
n_cluster = 3
X0 = np.random.multivariate_normal(means[0], cov, n_samples)
X1 = np.random.multivariate_normal(means[1], cov, n_samples)
X2 = np.random.multivariate_normal(means[2], cov, n_samples)
X = np.concatenate((X0, X1, X2), axis = 0)
X0
# len(X0)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'bo', markersize=5)
plt.plot()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D # 3d plot
from termcolor import colored as cl # text customization

from sklearn.preprocessing import StandardScaler # data normalization
from sklearn.cluster import KMeans # K-means algorithm
clusters = 3
model = KMeans(n_clusters = clusters)
model.fit(X)

labels = model.labels_
print(cl(labels[:500], attrs = ['bold']))

```

```python
print(model.labels_)
```

```python
model = KMeans(init = 'k-means++', n_clusters = clusters, n_init = 12)
```

```python
model.fit(X)

labels = model.labels_
print(cl(labels[:100], attrs = ['bold']))
```

```python
centroids = model.cluster_centers_

# create a scatter plot of the data points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)

# mark the centroids with red circles
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, linewidths=3, color='red', zorder=10)

plt.show()
```
