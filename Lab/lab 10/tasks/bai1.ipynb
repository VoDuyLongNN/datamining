```python
import numpy
import matplotlib.pyplot as plt
from termcolor import colored
from sklearn.cluster import KMeans
```

```python
c1 = numpy.array([2, 10])
c2 = numpy.array([2, 2])
c3 = numpy.array([10, 1])
c4 = numpy.array([10, 10])

# Khởi tạo ma trận dữ liệu bằng cách ghép các cụm lại với nhau
data = numpy.concatenate((c1.reshape(2,1), c2.reshape(2,1), c3.reshape(2,1)), axis=1)

# # Tính ma trận hiệp phương sai của dữ liệu
# cov = numpy.cov(data)

# # In ma trận hiệp phương sai
# print(cov)

cov = [[2, 0], [0, 2]]
```

```python
samples = 500
cluster = 4

X0 = numpy.random.multivariate_normal(c1, cov, samples)
X1 = numpy.random.multivariate_normal(c4, cov, samples)
X2 = numpy.random.multivariate_normal(c2, cov, samples)
X3 = numpy.random.multivariate_normal(c3, cov, samples)
X4 = numpy.random.multivariate_normal(c4, cov, samples)

X = numpy.concatenate((X0,X1,X2,X3,X4), axis=0)
X
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'o', color='#EAB543', markersize=5)
plt.plot()
plt.show()
```

```python
model = KMeans(init='k-means++',n_clusters=cluster, n_init=10)
model.fit(X)

labels = model.labels_
print(colored(labels[:500], attrs=['bold']))
```

```python
centroids = model.cluster_centers_

# create a scatter plot of the data points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)

# mark the centroids with red circles
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, linewidths=3, color='red', zorder=10)

plt.show()
```
