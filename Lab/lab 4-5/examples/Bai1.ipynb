```python
import pandas
import numpy

df = pandas.read_excel('../JOB_INFO.xlsx')
df
```

```python
df.isnull().sum()
```

```python
# Hiển thị những dòng có giá trị thiếu
df[df.isnull().any(axis=1)]
```

```python
# hiển thị những cột có giá trị thiếu và tổng số giá trị thiếu trên cột đó
for i in df:
   if df[i].isnull().any():
      print(i,': ',df[i].isnull().sum())
```

```python
df.dropna()
```

```python
df2 = df.copy()
df2['Age'].fillna(df2['Age'].mean().round(1), inplace=True)
df2
```

```python
df2.groupby(["Skill"])["Dept"].describe()
```

```python
dept_mode = df2.groupby('Skill')['Dept'].apply(lambda x: x.mode()[0]).to_dict()
dept_mode
df2['Dept'] = df2.apply(lambda x: dept_mode[x['Skill']] if pandas.isnull(x['Dept']) else x['Dept'], axis=1)
```

```python
df2[['ENO', 'Dept']]
```

```python
df2.to_excel('Clean_job_info.xlsx')
```
