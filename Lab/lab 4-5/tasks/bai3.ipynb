```python
import pandas 

df = pandas.read_csv('TotnghiepOK.csv')

df[['Toan', 'Anh']].isnull().sum()
```

```python
# Rời rạc hóa môn toán
toan_cut_labels = ['Yếu','Trung bình','Khá','Giỏi']
toan_bins = [0,4,6,8,10]
df['Toan'] = pandas.cut(df['Toan'],labels=toan_cut_labels, bins=toan_bins)
```

```python
# # Rời rạc hóa môn anh
Toan_cut_labels = ['Trung bình','Khá','Giỏi']
Toan_bins = [0,4,8,10]
df['Anh'] = pandas.cut(df['Anh'],labels=Toan_cut_labels, bins=Toan_bins)
```
