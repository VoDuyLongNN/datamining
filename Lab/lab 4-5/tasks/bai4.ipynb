```python
import pandas
import seaborn

df = pandas.read_csv('../105.csv')

df
```

```python
# Rời rạc hóa các trường thành 0,1 (0 cho những giá trị điểm >=6, 0 cho những giá trị còn lại)
cut_labels = [0,1]
cut_bins = [0,6,10]

for col in df:
   df[col] = pandas.cut(df[col], labels=cut_labels, bins=cut_bins)
df
```

```python
df.to_csv('105moi.csv')
```
