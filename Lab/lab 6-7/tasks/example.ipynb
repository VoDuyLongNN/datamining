```python
import pandas
from  apyori import apriori

data = pandas.read_csv('../105.csv')
data
```

```python
data.head(10)
```

```python
records = []
for i in range(0,99):
   records.append([str(data.values[i,j]) for j in range(0,10)])
records
```

```python
rules = apriori(records, min_support=0.01, min_confidence=0.3, min_lift=3, min_length=2)
rules_result = list(rules)
rules_result
```

```python
for item in rules_result:
   pair = item[0]
   items = [x for x in pair]
   print('rule: ' + items[0] + '->' + items[1])
   print('suport: '+ str(item[1]))
   print('confidence:'+str(item[2][0][2]))
   print('lift:'+str(item[2][0][3]))
   print('================================================')
```
